```python
from matplotlib import pyplot as plt
```

### Явная схема Эйлера 

```python
ar1 = []
ar2 = []
ar3 = []
with open("data_explicit_euler.txt", "r") as f:
    s = f.readlines()
for elem in s:
    a, b, c = elem.split()
    ar1.append(float(a))
    ar2.append(float(b))
    ar3.append(float(c))
    
plt.plot(ar1, ar2, 'r', label = 'voltage') # по косинусу колебания 
plt.plot(ar1, ar3, 'c', label = 'current') # по синусу колебания 
plt.xlabel('Время, c')
plt.ylabel('U_C / I, В/A')
plt.legend()
plt.grid()
plt.show()
```

### Неявная схема Эйлера 

```python
ar4 = []
ar5 = []
ar6 = []
with open("data_implicit_euler.txt", "r") as f:
    s = f.readlines()
    #print(s)
for elem in s:
    a, b, c = elem.split()
    ar4.append(float(a))
    ar5.append(float(b))
    ar6.append(float(c))

plt.plot(ar4, ar5, 'r', linewidth = 1.5, label = 'voltage') # по косинусу колебания 
plt.plot(ar4, ar6, 'c', linewidth = 1.5, label = 'current') # по синусу колебания 
plt.xlabel('Время, c')
plt.ylabel('U_C / I, В/A')
plt.legend(loc = 'upper right')
plt.grid()
plt.show()    
```

### Метод Рунге-Кутты четвертого порядка 

```python
ar7 = []
ar8 = []
ar9 = []
with open("data_runge_kutta.txt", "r") as f:
    s = f.readlines()
for elem in s:
    a, b, c = elem.split()
    ar7.append(float(a))
    ar8.append(float(b))
    ar9.append(float(c))

plt.plot(ar7, ar8, "r", linewidth = 1.3, label = 'Зависимость напряжения от времени')
plt.plot(ar7, ar9, "c", linewidth = 1.3, label = 'Зависимость тока от времени')
plt.xlabel('Время, c')
plt.ylabel('U_C / I, В/A')
plt.legend(loc = 'upper left')
plt.grid()
plt.show()   
```

### Ошибка численного решения

```python
import numpy as np 
```

#### Cетка 2000 шагов

```python
t_anal = ar7
y_anal = np.cos(t_anal)
t_rk = ar7
y_rk = ar8
plt.plot(t_rk, y_anal-y_rk, 'c', linewidth = 0.5)
plt.title('Ошибка численного решения (2000 шагов)')
plt.xlabel('Время, с')
plt.ylabel('Delta U_C, В')
plt.grid()
plt.show()
```

#### Cетка 4000 шагов

```python
aar1 = []
aar2 = []
aar3 = []
with open("data_4000_rk.txt", "r") as f:
    s = f.readlines()
for elem in s:
    a, b, c = elem.split()
    aar1.append(float(a))
    aar2.append(float(b))
    aar3.append(float(c))

t_anal = aar1
y_anal = np.cos(t_anal)
t_rk = aar1
y_rk = aar2
plt.plot(t_anal, y_anal-y_rk, 'r', linewidth = 0.5)
plt.title('Ошибка численного решения (4000 шагов)')
plt.xlabel('Время, с')
plt.ylabel('Delta U_C, В')
plt.grid()
plt.show()
```
